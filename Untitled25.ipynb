```python

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
X_audio = X.astype('float32')#Cast all numerical features to float32 to reduce memory usage
```

```python
df['title_clean'] = (
    Y['about_song']
    .str.lower()
    .str.replace(r'[^a-z0-9 ]', '', regex=True))
```

```python
# it removes all the common words which are repeated most of the time and low information words
tfidf = TfidfVectorizer(
    stop_words='english',
    ngram_range=(1, 2),
    max_features=5000  #reducing the words and only  use top 5000 most important word ,its reduces the load on the ram and make workflow smooth.
)

X_title = tfidf.fit_transform(df['title_clean'])
```

```python
indices = pd.Series(Y.index, index=Y['about_song']).drop_duplicates()#its create the dataframe and provide row number to every song and remove duplicate.
```

```python
def audio_similarity_one(idx, X_audio):
    return cosine_similarity(X_audio[idx:idx+1], X_audio).flatten()
# This function computes similarity based only on numeric audio features,
# since cosine similarity operates on numerical vectors
```

```python
def title_similarity_one(idx, X_title):
    return cosine_similarity(X_title[idx], X_title).flatten()
# Compute cosine similarity between one song title and all others
# using TF-IDF text feature vectors
```

```python
def hybrid_similarity_one(
    idx,
    X_audio,
    X_title,
    alpha=0.8
):
    sim_audio = audio_similarity_one(idx, X_audio)
    sim_title = title_similarity_one(idx, X_title)

    return alpha * sim_audio + (1 - alpha) * sim_title
# Hybrid similarity combining audio-based and text-based similarity
# Audio features (e.g., loudness, energy, acousticness, valence) are given higher weight (80%)
# since they better represent how a song sounds, while TF-IDF title similarity contributes 20%
```

```python
# User inputs a song title
# Title → index mapping
# Compute hybrid similarity
# Rank songs by similarity
# Remove self-match
# Apply similarity threshold
# Return top-N recommendations

def recommend_hybrid(
    song_title,
    df,
    X_audio,
    X_title,
    n=5,
    min_sim=0.6,
    alpha=0.8
):

    idx = indices[song_title]


    sim_scores = hybrid_similarity_one(
        idx, X_audio, X_title, alpha
    )


    sim_scores = list(enumerate(sim_scores))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)


    sim_scores = [
        (i, score) for i, score in sim_scores[1:]
        if score >= min_sim
    ][:n]

    song_indices = [i[0] for i in sim_scores]

    return Y['about_song'].iloc[song_indices].values
```

```python
random_song =Y['about_song'][496]

recommend_hybrid(
    song_title=random_song,
    df=Y,
    X_audio=X_audio,
    X_title=X_title,
    n=5,
    alpha=0.8
)
```
